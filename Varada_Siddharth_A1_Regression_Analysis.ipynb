```python
# importing libraries
import pandas as pd # data science essentials
import matplotlib.pyplot as plt # essential graphical output
import seaborn as sns # enhanced graphical output

file = './birthweight_low.xlsx'


# reading the file into Python
birth_weight = pd.read_excel(io    = file,
                        header     = 0,
                       sheet_name  = 0)


# outputting the first ten rows of the dataset
birth_weight.head(n = 10)
```

```python
#base model with original continuous variables
# importing an additional package
import statsmodels.formula.api as smf # predictive modeling with nice outputs
# Step 1: INSTANTIATE a model object
lm_best = smf.ols(formula =  """bwght ~      mage +
                                             fage +
                                             npvis +
                                             meduc +
                                             feduc +
                                             cigs + 
                                             drink""",
                                data = birth_weight)


# Step 2: FIT the data into the model object
results = lm_best.fit()


# Step 3: analyze the SUMMARY output
print(results.summary())
```

```python
# taking the birth weight dataset
# and then
# transforming it into boolean based on if a value is null
# and then
# summing together the results per column

birth_weight.isnull().sum(axis = 0)
```

```python
# Calculating the median and checking if sample size is suitable for grouping

median_meduc = birth_weight['meduc'].median()
median_feduc = birth_weight['feduc'].median()
median_cigs = birth_weight['cigs'].median()
median_drink = birth_weight['drink'].median()
median_npvis = birth_weight['npvis'].median()

# splitting the variables based on conditions
meduc_high   = len(birth_weight['meduc'][birth_weight['meduc'] > median_meduc])
feduc_high   = len(birth_weight['feduc'][birth_weight['feduc'] > median_feduc])
cigs_high    = len(birth_weight['cigs'][birth_weight['cigs'] > median_cigs])
drink_high   = len(birth_weight['drink'][birth_weight['drink'] > median_drink])


## the table below has been coded for you ##

# printing a table of the results
print(f"""
                         Yes\t\tNo
                       ---------------------
Mother education        | {meduc_high}\t\t{len(birth_weight) - meduc_high}
Father education        | {feduc_high}\t\t{len(birth_weight) - feduc_high}
Cigarettes              | {cigs_high}\t\t{len(birth_weight) - cigs_high}
Drinks                  | {drink_high}\t\t{len(birth_weight) - drink_high}
# """)
# print(mean_meduc)
# print(mean_feduc)
# print(mean_cigs)
# print(mean_drink)
```

```python
# instantiating imputation values 
meduc_fill = median_meduc
feduc_fill = median_feduc
npvis_fill = median_npvis

# imputing 'meduc'
birth_weight['meduc'] = birth_weight['meduc'].fillna(value = meduc_fill)
# imputing 'feduc'
birth_weight['feduc'] = birth_weight['feduc'].fillna(value = feduc_fill)
# npvis
birth_weight['npvis'] = birth_weight['npvis'].fillna(value = npvis_fill)
```

```python
#checking if null values are filled
birth_weight.isnull().any(axis = 0).any(axis = 0)
```

```python
# placeholder variables
birth_weight['high_meduc']  = 0
birth_weight['high_feduc'] = 0
birth_weight['heavy_smoker']  = 0
birth_weight['heavy_drinker'] = 0

#new binary variables which identify with degree of education and addiction
for index, value in birth_weight.iterrows():
    

    # meduc
    if birth_weight.loc[index, 'meduc'] > 13.0:
        birth_weight.loc[index, 'high_meduc'] = 1
        
    # feduc
    if birth_weight.loc[index, 'feduc'] > 14.0:
        birth_weight.loc[index, 'high_feduc'] = 1
    
    # cigs
    if birth_weight.loc[index, 'cigs'] > 11.0:
        birth_weight.loc[index, 'heavy_smoker'] = 1
        
        
    # drink
    if birth_weight.loc[index, 'drink'] > 5.0:
        birth_weight.loc[index, 'heavy_drinker'] = 1
```

```python
# checking results
birth_weight[  ['high_meduc', 'high_feduc', 'heavy_smoker', 'heavy_drinker']  ].head(n = 5)
```

```python
#Transforming variables to get a improve linearity and get a better fit in the model
import numpy as np
birth_weight['log_meduc'] = np.log(birth_weight['meduc'] + 0.001)
birth_weight['log_feduc'] = np.log(birth_weight['feduc'] + 0.001)
birth_weight['log_bwght'] = np.log(birth_weight['bwght'] + 0.001)
birth_weight['log_monpre'] = np.log(birth_weight['monpre'])
birth_weight['log_mage'] = np.log(birth_weight['mage'])
birth_weight['log_fage'] = np.log(birth_weight['fage'])
birth_weight['log_npvis'] = np.log(birth_weight['npvis'] + 0.001)
birth_weight['log_cigs'] = np.log(birth_weight['cigs'] + 0.001)
birth_weight['log_drink'] = np.log(birth_weight['drink'] + 0.001)

#correlations between different versions of variables
log_corr = birth_weight.loc[ : , ['meduc',
                             'log_meduc',
                             'feduc',
                             'log_feduc', 'monpre', 'log_monpre', 'npvis', 
                            'log_npvis', 'mage', 'log_mage', 'fage',
                            'log_fage', 'cigs', 'log_cigs', 'drink', 'log_drink',
                                 'bwght', 'log_bwght']  ].corr(method = 'pearson')\
                                                 .round(decimals = 2)



log_corr.loc[ ['meduc', 'log_meduc', 'feduc',
                             'log_feduc', 'monpre', 'log_monpre', 'npvis', 
                            'log_npvis', 'mage', 'log_mage', 'fage',
                            'log_fage', 'cigs', 'log_cigs', 'drink', 'log_drink'],
              ['bwght', 'log_bwght'] ]
```

```python
#combining cigarette consumption and drink consumption into a new feature
birth_weight["risk"] = birth_weight["cigs"] + birth_weight["drink"] 
```

```python
#Grouping of races into further segments
birth_weight["mnonwhte"]=0
for index, value in birth_weight.iterrows():
    # mnonwhte
    if birth_weight.loc[index, 'mwhte'] == 0:
        birth_weight.loc[index, 'mnonwhte'] = 1
birth_weight["fnonwhte"]=0
for index, value in birth_weight.iterrows():
    # fnonwhte
    if birth_weight.loc[index, 'fwhte'] == 0:
        birth_weight.loc[index, 'fnonwhte'] = 1
birth_weight["heavy_both"]=0
for index, value in birth_weight.iterrows():
    # heavy_both
    if birth_weight.loc[index, 'heavy_smoker'] == 1 & birth_weight.loc[index, 'heavy_drinker'] == 1:
        birth_weight.loc[index, 'heavy_both'] = 1
birth_weight["is_2nd_monpre"]=0
for index, value in birth_weight.iterrows():
    # is_2nd_monpre
    if birth_weight.loc[index, 'monpre'] == 2.0:
        birth_weight.loc[index, 'is_2nd_monpre'] = 1
birth_weight["par_whte"]=0
for index, value in birth_weight.iterrows():
    # par_whte
    if birth_weight.loc[index, 'mwhte'] == 1 & birth_weight.loc[index, 'fwhte'] == 1:
        birth_weight.loc[index, 'par_whte'] = 1
birth_weight["par_blck"]=0
for index, value in birth_weight.iterrows():
    # par_blck
    if birth_weight.loc[index, 'mblck'] == 1 & birth_weight.loc[index, 'fblck'] == 1:
        birth_weight.loc[index, 'par_blck'] = 1
birth_weight["par_oth"]=0
for index, value in birth_weight.iterrows():
    # par_oth
    if birth_weight.loc[index, 'moth'] == 1 & birth_weight.loc[index, 'foth'] == 1:
        birth_weight.loc[index, 'par_oth'] = 1
birth_weight["par_nonwhte"]=0
for index, value in birth_weight.iterrows():
    # par_nonwhte
    if birth_weight.loc[index, 'par_blck'] == 1 or birth_weight.loc[index, 'par_oth'] == 1:
        birth_weight.loc[index, 'par_nonwhte'] = 1
```

```python
#Base model to send to testing
# Step 1: INSTANTIATE a model object
lm_best = smf.ols(formula = """log_bwght ~ mage +
                                       meduc + 
                                       fage +
                                       feduc +
                                       cigs +
                                       drink +
                                       mwhte +
                                       mblck +
                                       moth +
                                       fwhte +
                                       fblck +
                                       foth +
                                       mnonwhte +
                                       fnonwhte +
                                       par_whte +
                                       par_blck +
                                       par_oth +
                                       log_npvis +
                                       log_monpre +
                                       high_meduc +
                                       high_feduc +
                                       heavy_smoker +
                                       heavy_drinker +
                                       heavy_both +
                                       risk""",
                  data = birth_weight)


# Step 2: FIT the data into the model object
results = lm_best.fit()


# Step 3: analyze the SUMMARY output
print(results.summary())
```

```python
from sklearn.model_selection import train_test_split # train/test split
from sklearn.linear_model import LinearRegression # linear regression (scikit-learn)

#OLS Model
# preparing explanatory variable data
birth_weight_data   = birth_weight.drop(["bwght",
                               "log_bwght",
                               "mage", "log_monpre","meduc","feduc",
                                "log_npvis", "cigs", "drink", "log_drink", 
                                         "high_meduc", "heavy_drinker", 
                                         "omaps", "fmaps", "male", 
                                         "high_feduc", "log_meduc", "log_mage",
                                          "log_fage",
                                        "is_2nd_monpre", "par_blck", "par_nonwhte"],
                               axis = 1)


# preparing response variables
birth_weight_target = birth_weight.loc[ : , "bwght"]
log_birth_weight_target = birth_weight.loc[ : , "log_bwght"]


# preparing training and testing sets (all letters are lowercase)
x_train, x_test, y_train, y_test = train_test_split(
            birth_weight_data,
            log_birth_weight_target,
            test_size = 0.25,
            random_state = 219)


# checking the shapes of the datasets
print(f"""
Training Data
-------------
X-side: {x_train.shape}
y-side: {y_train.shape}


Testing Data
------------
X-side: {x_test.shape}
y-side: {y_test.shape}
""")
```

```python
# merging X_train and y_train so that they can be used in statsmodels
birth_weight_train = pd.concat([x_train, y_train], axis = 1)

#buidling the model after removing insignificant variables
# Step 1: build a model
lm_best = smf.ols(formula =  """log_bwght ~ fage +
                                       npvis +
                                       monpre +
                                       mwhte +
                                       mblck +
                                       moth +
                                       fwhte +
                                       fblck +
                                       foth +
                                       mnonwhte +
                                       fnonwhte +
                                       par_whte +
                                       par_oth +
                                       log_feduc +
                                       log_cigs +
                                       heavy_both +
                                       risk""",
                                data = birth_weight_train)


# Step 2: fit the model based on the data
results = lm_best.fit()



# Step 3: analyze the summary output
print(results.summary())
```

```python
# applying modelin scikit-learn

# preparing x-variables from the OLS model
ols_data = birth_weight.loc[ : , ["fage", "npvis", "monpre", "mwhte",
                                       "mblck",
                                       "moth",
                                       "fwhte",
                                       "fblck",
                                       "foth",
                                       "mnonwhte",
                                       "fnonwhte",
                                       "par_whte",
                                       "par_oth",
                                       "log_feduc",
                                       "log_cigs",
                                       "heavy_both",
                                       "risk"]]


# preparing response variable
log_birth_weight_target = birth_weight.loc[ : , 'log_bwght']


###############################################
## setting up more than one train-test split ##
###############################################
# FULL X-dataset (normal Y)
x_train_FULL, x_test_FULL, y_train_FULL, y_test_FULL = train_test_split(
            birth_weight_data,     # x-variables
            log_birth_weight_target,   # y-variable
            test_size = 0.25,
            random_state = 219)


# OLS p-value x-dataset (normal Y)
x_train_OLS, x_test_OLS, y_train_OLS, y_test_OLS = train_test_split(
            ols_data,         # x-variables
            log_birth_weight_target,   # y-variable
            test_size = 0.25,
            random_state = 219)
```

```python
# INSTANTIATING a model object
lr = LinearRegression()


# FITTING to the training data
lr_fit = lr.fit(x_train_OLS, y_train_OLS)


# PREDICTING on new data
lr_pred = lr_fit.predict(x_test_OLS)


# SCORING the results
print('OLS Training Score :', lr.score(x_train_OLS, y_train_OLS).round(4))  # using R-square
print('OLS Testing Score  :',  lr.score(x_test_OLS, y_test_OLS).round(4)) # using R-square

lr_train_score = lr.score(x_train_OLS, y_train_OLS).round(4)
lr_test_score = lr.score(x_test_OLS, y_test_OLS).round(4)

# displaying and saving the gap between training and testing
print('OLS Train-Test Gap :', abs(lr_train_score - lr_test_score).round(4))
lr_test_gap = abs(lr_train_score - lr_test_score).round(4)
```

```python
#ARD Model
import sklearn.linear_model
# INSTANTIATING a model object
ard_model = sklearn.linear_model.ARDRegression(normalize  = False)


# FITTING the training data
ard_fit = ard_model.fit(x_train_FULL, y_train_FULL)


# PREDICTING on new data
ard_pred = ard_fit.predict(x_test_FULL)


print('Training Score:', ard_model.score(x_train_FULL, y_train_FULL).round(4))
print('Testing Score :',  ard_model.score(x_test_FULL, y_test_FULL).round(4))


# saving scoring data for future use
ard_train_score = ard_model.score(x_train_FULL, y_train_FULL).round(4)
ard_test_score  = ard_model.score(x_test_FULL, y_test_FULL).round(4)


# displaying and saving the gap between training and testing
print('ARD Train-Test Gap :', abs(ard_train_score - ard_test_score).round(4))
ard_test_gap = abs(ard_train_score - ard_test_score).round(4)
```
